```python
import json

geo_path = 'hangjungdong.json'
geo_json = json.load(open(geo_path, encoding="utf-8"))

# 행정동 json 파일 로드
```

```python
geo_json['features'][0]
# 확인
```

```python
def geo(data,lst_all):
    lst_all = []
    for i in range(len(data['features'])):
        if data['features'][i]['properties']['adm_cd8'][0:2] == '11':
            lst_all.append(data['features'][i])
    return lst_all
        
lst_all = geo(geo_json,lst_all)
# Geojson파일에서, 필요한 데이터 만을 list로 저장
```
